```python

topic_data = "원본데이터/주제별 일상 대화 데이터/"
```

```python
topic = os.listdir("원본데이터/주제별 일상 대화 데이터/")
```

```python
topic_ = []
for i in topic:
    target = os.listdir(topic_data+'/'+i)
    for j in range(len(target)):
        path = topic_data+'/'+str(i)+'/'+str(target[j])
        # print(target[j])
        try:
            ex1 = open(path, encoding="UTF-8")
            ex1= json.loads(ex1.read())
            for k in range(len(ex1['info'][0]['annotations']['lines'])):
                    topic_.append(ex1['info'][0]['annotations']['lines'][k]['norm_text'])
                    print(topic_)
        except:
            print('Error')
```
